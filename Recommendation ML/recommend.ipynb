```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies_metadata.csv')
credits = pd.read_csv('credits.csv') 
keywords = pd.read_csv('keywords.csv') 
```

```python
movies.head()
```

```python
movies = movies[['id','original_language','title','overview','genres']]
```

```python
movies=movies[movies['original_language']=='en']
```

```python
movies=movies.merge(keywords,on='id')
```

```python
movies
```

```python
movies.dropna()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
from nltk.tokenize import word_tokenize
```

```python
movies['overview']=movies['overview'].apply(str)
```

```python
movies['overview']=movies['overview'].apply(word_tokenize)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
final = movies.drop(columns=['overview','genres','keywords','cast','crew','original_language'])
```

```python
final
```

```python
final['tags'] = final['tags'].apply(lambda x: " ".join(x))
final.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')

vector = cv.fit_transform(final['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = final[final['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:15]:
        print(final.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
